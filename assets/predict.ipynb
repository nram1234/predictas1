```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
df=pd.read_csv("modd.csv")
df.head()
```

```python
#Deleting useless column 
df=df.drop("ID", axis=1)
df.head()
```

```python
df['Year'] = df.Date.str.extract(r'\b(\w+)$', expand=True)
df['Year']= df['Year'].astype(int)
df['Code_departement']= df['Code_departement'].astype(str).astype(int)
df=df.drop("Date", axis=1)
df_x = df[['Year','Code_departement','Code_Commune','Code_region']]
df_y = df['prixm2']
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test = train_test_split(df_x,df_y,test_size=0.2,random_state=0) 
reg = LinearRegression() 
df_x.describe()
```

```python
reg.fit(x_train,y_train)
```

```python
print(reg.coef_)
```

```python
y_pred = reg.predict(x_test)
print(y_pred)
```

```python
print(y_test)
```

```python
def guess (model,year,dep,com,reg):
  x=np.array([year+4,dep,com,reg]).reshape(1,4)
  return(model.predict(x))
df['price_predicted']=0.0
df.head()
for i in range(len(df['Year'])):
  df['price_predicted'][i]=guess(reg,df['Year'][i],df['Code_Commune'][i],df['Code_departement'][i],df['Code_region'][i])
```

```python
for i in range(len(df['Year'])):
  if df['price_predicted'][i]<0:
    df['price_predicted'][i]=-df['price_predicted'][i]
```
